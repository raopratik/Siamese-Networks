```python
X_1 = np.load('siamese_travel_ids.npy')
X_2 = np.load('siamese_travel_ids1.npy')
```

```python
from sklearn.model_selection import train_test_split
X_1_train,X_1_test,X_2_train,X_2_test,y_train,y_test = train_test_split(X_1,X_2,data[['y','y1']], test_size=.2)
```

```python
def getTrainBatch(j):
    i = j*batchSize
    x_1 = X_1_train[i:i+batchSize]
    x_2 = X_2_train[i:i+batchSize]
    y = y_train[['y','y1']][i:i+batchSize]
    return x_1, x_2, y
```

```python
def getTestBatch(j):
    i = j*batchSize
    x_1 = X_1_test[i:i+batchSize]
    x_2 = X_2_test[i:i+batchSize]
    y = y_test[['y','y1']][i:i+batchSize]
    return x_1, x_2, y
```

```python
import tensorflow as tf
tf.reset_default_graph()
labels = tf.placeholder(tf.float32, [batchSize, numClasses])
input_data1 = tf.placeholder(tf.int32, [batchSize, maxSeqLength])
input_data2 = tf.placeholder(tf.int32, [batchSize, maxSeqLength])
```

```python
wordVectors = tf.convert_to_tensor(wordVectors, dtype=tf.float32)
```

```python
data1 = tf.placeholder(shape = (batchSize,maxSeqLength,numDimensions), dtype=tf.float32)
data1 = tf.nn.embedding_lookup(wordVectors,input_data1)
data2 = tf.placeholder(shape = (batchSize,maxSeqLength,numDimensions), dtype=tf.float32)
data2 = tf.nn.embedding_lookup(wordVectors,input_data2)
```

```python
fw_cell1 = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
bw_cell1 = tf.nn.rnn_cell.MultiRNNCell([ tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
fw_cell2 = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
bw_cell2 = tf.nn.rnn_cell.MultiRNNCell([ tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
```

```python
outputs1,value1 = tf.nn.bidirectional_dynamic_rnn(fw_cell1, bw_cell1,data1,dtype=tf.float32,scope='biLSTM1')
outputs2,value2 = tf.nn.bidirectional_dynamic_rnn(fw_cell2, bw_cell2,data2,dtype=tf.float32,scope='biLSTM2')
```

```python
value1 = tf.concat(value1, 3)
value2 = tf.concat(value2, 3)
```

```python
weight = tf.Variable(tf.truncated_normal([100, numClasses]))
bias = tf.Variable(tf.constant(0.1, shape=[numClasses]))
```

```python
prediction = (tf.matmul(dense_values, weight) + bias)
correctPred = tf.equal(tf.argmax(prediction,1), tf.argmax(labels,1))
accuracy = tf.reduce_mean(tf.cast(correctPred, tf.float32))
t = tf.argmax(labels,1)
tt = tf.argmax(prediction,1)
```

```python
import datetime

tf.summary.scalar('Loss', loss)
tf.summary.scalar('Accuracy', accuracy)
merged = tf.summary.merge_all()
logdir = "tensorboard/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + "/"
writer = tf.summary.FileWriter(logdir, sess.graph)
```

```python
sess = tf.InteractiveSession()
saver = tf.train.Saver()
sess.run(tf.global_variables_initializer())

for i in range(epochs):
    print("epoch:  "+str(i))
    #Next Batch of reviews
    l2 = 0
    acc_ = []
    val_ = []
    
    for j in range(len(X_1_train)//batchSize):
        nextBatch1, nextBatch2, nextBatchLabels = getTrainBatch(j);
        #print((j*batchSize*100)/len(X_1))
        acc,l,_ = sess.run([accuracy,loss,optimizer], {input_data1: nextBatch1,input_data2: nextBatch2, labels: nextBatchLabels})
        nextBatcht1, nextBatcht2, nextBatchLabelst = getTestBatch(3);
        #print((j*batchSize*100)/len(X_1))
        val = sess.run(accuracy, {input_data1: nextBatcht1,input_data2: nextBatcht2, labels: nextBatchLabelst})
        l2 = l2 + l
        acc_.append(acc)
        val_.append(val)
        
            
    
    
    #Write summary to Tensorboard
        if (j % 160 == 0):
            summary = sess.run(merged, {input_data1: nextBatch1,input_data2: nextBatch2, labels: nextBatchLabels})
            writer.add_summary(summary, j)

        #Save the network every 10,000 training iterations
        if (j % 50 == 0 and j != 0):
            save_path = saver.save(sess, "models/pretrained_lstm.ckpt", global_step=i)
            print("saved to %s" % save_path)
    print('accuracy mean: ',np.array(acc_).sum()/len(acc_), ' val mean:', np.array(val_).sum()/len(val_))
    print("epoch loss: ",l2)
writer.close()
```
