```python
fw_cell1 = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(4)])
bw_cell1 = tf.nn.rnn_cell.MultiRNNCell([ tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(4)])
fw_cell2 = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(4)])
bw_cell2 = tf.nn.rnn_cell.MultiRNNCell([ tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(4)])
```

```python
import datetime

tf.summary.scalar('Loss', loss)
tf.summary.scalar('Accuracy', accuracy)
merged = tf.summary.merge_all()
logdir = "tensorboard/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + "/"
writer = tf.summary.FileWriter(logdir, sess.graph)
```

```python
sess = tf.InteractiveSession()
saver = tf.train.Saver()
sess.run(tf.global_variables_initializer())

for i in range(epochs):
    print("epoch:  "+str(i))
    #Next Batch of reviews
    for j in range(len(X_1)//batchSize):
        nextBatch1, nextBatch2, nextBatchLabels = getTrainBatch(j);
        print((j*batchSize*100)/len(X_1))
        acc,l,_ = sess.run([accuracy,loss,optimizer], {input_data1: nextBatch1,input_data2: nextBatch2, labels: nextBatchLabels})
        print("accuracy: ",acc,' loss: ',l)
    #Write summary to Tensorboard
        if (j % 5 == 0):
            summary = sess.run(merged, {input_data1: nextBatch1,input_data2: nextBatch2, labels: nextBatchLabels})
            writer.add_summary(summary, j)

        #Save the network every 10,000 training iterations
        if (j % 50 == 0 and j != 0):
            save_path = saver.save(sess, "models/pretrained_lstm.ckpt", global_step=i)
            print("saved to %s" % save_path)
writer.close()
```
