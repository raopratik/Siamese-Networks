```python
fw_cell1 = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
bw_cell1 = tf.nn.rnn_cell.MultiRNNCell([ tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
fw_cell2 = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
bw_cell2 = tf.nn.rnn_cell.MultiRNNCell([ tf.nn.rnn_cell.BasicLSTMCell(lstmUnits, state_is_tuple=True) for _ in range(2)])
```

```python
outputs1,value1 = tf.nn.bidirectional_dynamic_rnn(fw_cell1, bw_cell1,data1,dtype=tf.float32,scope='biLSTM1')
outputs2,value2 = tf.nn.bidirectional_dynamic_rnn(fw_cell2, bw_cell2,data2,dtype=tf.float32,scope='biLSTM2')
```

```python
value1 = tf.concat(value1, 3)
value2 = tf.concat(value2, 3)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    x_1, x_2, y = getTrainBatch(1)
    val1, val2 = sess.run([value1,value2], feed_dict={input_data1: x_1,input_data2: x_2, labels: y })
    print(val1[0][0].shape,val2[0][0].shape)
```

```python
    # Euclidean distance between x1,x2
l2diff = tf.sqrt( tf.reduce_sum(tf.square(tf.subtract(value1, value2)),
                                    reduction_indices=1))

    # you can try margin parameters
margin = tf.constant(1.)     

labels = tf.to_float(labels)

match_loss = tf.square(l2diff, 'match_term')
mismatch_loss = tf.maximum(0., tf.subtract(margin, tf.square(l2diff)), 'mismatch_term')

    # if label is 1, only match_loss will count, otherwise mismatch_loss
loss = tf.add(tf.multiply(labels, match_loss),tf.multiply((1 - labels), mismatch_loss), 'loss_add')

loss_mean = tf.reduce_mean(loss)
```

```python
optimizer = tf.train.AdamOptimizer().minimize(loss_mean)
```
